```python
weights = pd.read_csv("cooccurrence_weights.csv")
weights.to_csv("cooccurrence_weights.csv.gz", index=False)
```

```python
counts = np.bincount(weights.values.flatten(), minlength=int(weights.values.flatten().max()))
probabilities = counts / weights.size

# We skip the 0-th
probabilities = probabilities[1:]

log_prob = np.log(probabilities)

- np.sum(probabilities * log_prob)
```

```python
(counts / weights.size), counts
```

```python
weights.hist(log=True, bins=10)
plt.title("Weights Distribution of Keys & Pentagrams")
plt.xlabel("Weights (linear scale)")
plt.ylabel("Frequency (log scale)")
plt.tight_layout()
```
